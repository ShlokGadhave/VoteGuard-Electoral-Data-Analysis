Feature Engineering 

```python


bins = [18, 25, 40, 60, 100]

labels = ['18-25', '26-40', '41-60', '60+']

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

print(df[['age', 'age_group']].head())
```

```python
#Creation of registration year column

df['registration_date'] = pd.to_datetime(df['registration_date'])
df['registration_year'] = df['registration_date'].dt.year
print(df[['registration_date', 'registration_year']].head())
```

```python
# Create a flag: missing_contact_info
df['missing_contact_info']=df["phone_number"].isnull()
print(df[['phone_number','missing_contact_info']].head())
```

```python
# Create duplicate Aadhaar flag.
df['Aadhaar Flag']=df.duplicated(subset='aadhaar_number')
print(df[['Aadhaar Flag','aadhaar_number']].head())
```

Visualizations

```python
#Voters per state 
sns.barplot(x='state',data=df,hue='gender')
plt.title('Voters per state')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gender Distribution
gender_counts = df['gender'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))

plt.pie(
    gender_counts.values,
    labels=gender_counts.index,
    shadow=True,
    autopct='%1.1f%%'
)

plt.title('Gender Distribution of Voters')

plt.show()
```

```python
sns.histplot(x='age',data=df,color='red')
plt.title('Age Disturbution by Voters')
plt.show()
```

```python
sns.countplot(x='registration_year',data=df)
plt.title('Registration Year By Voters')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_cities = df['city'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_cities.index,
    y=top_cities.values
)

plt.title('Top 10 Cities by Voter Count')

plt.xlabel('City')

plt.ylabel('Number of Voters')

plt.xticks(rotation=45)

plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.heatmap(
    df.isnull(),
    cbar=False,
    yticklabels=False
)

plt.title('Missing Values Heatmap')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='gender',
    y='age',
    data=df
)

plt.title('Age Distribution by Gender')

plt.xlabel('Gender')

plt.ylabel('Age')

plt.show()
```

```python
# Count duplicate Aadhaar cases by state
fraud_state = df[df['Aadhaar Flag'] == True]['state'].value_counts()

plt.figure(figsize=(12,6))

sns.barplot(
    x=fraud_state.index,
    y=fraud_state.values
)

plt.title('Fraud Distribution Across States')

plt.xlabel('State')

plt.ylabel('Duplicate Aadhaar Cases')

plt.xticks(rotation=45)

plt.show()
```

```python
# Count registrations by year
registration_trend = df['registration_year'].value_counts().sort_index()

plt.figure(figsize=(10,6))

plt.plot(
    registration_trend.index,
    registration_trend.values,
    marker='o'
)

plt.title('Voter Registrations Over Time')

plt.xlabel('Registration Year')

plt.ylabel('Number of Registrations')

plt.grid(True)

plt.show()
```

```python

df['fraud_score'] = (
    df['Aadhaar Flag'].astype(int)
)

# Risk Categories
df['risk_category'] = pd.cut(
    df['fraud_score'],
    bins=[-1, 0, 1],
    labels=['Low Risk', 'High Risk']
)

print(df[['fraud_score', 'risk_category']].head(10))
```
